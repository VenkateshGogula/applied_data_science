```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the datasets
# Loading the dataset
df = pd.read_csv("c/Documents/data science projects/assegments/datasets/StudentPerformanceFactors.csv")

# Display first few rows of the datasets
df.head(10)
```

```python
#size of the dataframe
print (df.shape)
```

```python
df.describe()
```

```python
# Display basic statistics
basic_stats = df[['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores', 'Exam_Score']].describe()
print(basic_stats)
```

```python


# Correlation matrix
correlation_matrix = df[['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores', 'Exam_Score']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

# scatter plot 

```python
sns.scatterplot(data=df, x='Hours_Studied', y='Exam_Score')
plt.title("Exam Score vs. Hours Studied")
plt.show()
```

# box plot 

```python
sns.boxplot(data=df, x='Motivation_Level', y='Exam_Score')
plt.title("Exam Score by Motivation Level")
plt.show()
```
